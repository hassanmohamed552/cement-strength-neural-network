part A

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

download the data

```python
concrete_data = pd.read_csv('https://cocl.us/concrete_data')# get the data
concrete_data.head() #display first 5 rows
```

split the data

```python
concrete_data_columns = concrete_data.columns

x = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
y = concrete_data['Strength'] # Strength column
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42) #split the data
n_cols = x.shape[1] # number of predictors
```

importing keras

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

model parameters

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

training the model

```python
# build the model
model = regression_model()
```

```python
# fit the model
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=50,verbose=2)
```

```python
y_pred=model.predict(X_test) #predicted values
scores = mean_squared_error(y_test, y_pred) #mean square error
```

```python
print(scores) #show the mean square error
```

repeating the splitting of data and the model 50 times

```python
score=[]

for i in range(1,51):
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42) #split the data
    n_cols = x.shape[1] # number of predictors
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2)
    y_pred=model.predict( X_test ) #predicted values
    scores_1 = mean_squared_error(y_test, y_pred) #mean square error
    score.append(scores_1)
```

mean and std

```python
mean=np.mean(score)#average of mean square errors
std=np.std(score)#std of mean squre errors
print("mean={},std={}".format(mean,std))
```
